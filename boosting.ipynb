```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for column in X.columns:
    if X[column].dtype not in ['int64', 'float64']:
        X[column] =le.fit_transform(X[column])
```

```python
X
```

```python
train = train.dropna(axis=0)
y = train['SalePrice']
X = train.drop('SalePrice', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

forest = RandomForestRegressor(n_estimators=800, n_jobs=-1).fit(X_train, y_train)
```

```python
forest
```

```python
preds = forest.predict(X_test)
```

```python
me = mean_absolute_error(y_test, preds)
me
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
boost = GradientBoostingRegressor(learning_rate=0.001, n_estimators=10_000).fit(X_train, y_train)
```

```python
b_preds = boost.predict(X_test)
```

```python
mean_absolute_error(y_test, b_preds)
```

```python
import numpy as np
def plot_feature_importances(model):
    plt.figure(figsize=(20, 40))
    n_features = X.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), X.columns)
    plt.xlabel("Importance")
    plt.ylabel("Feature")
```

```python
plot_feature_importances(boost)
```
